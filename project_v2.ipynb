## Борьба с дисбалансом

Как будем бороться с дисбалансом классов? Можно проверить два способа, 1) Взвесить классы при обучении модели 2) Изменить размер выборки (upsampling, downsampling). Начнем с первого способа.

### Взвешивание классов

#### Логистическая регрессия

### Upsampling

Создадим функцию, которая сама увеличит выборку в нужное количество нам раз. На этапе подготовки данных мы поняли, что ответов отрицательного класса в нашем датасете примерно в 3.5 раза больше, чем ответов положительного класса. Увеличим нашу число ответов положительного класса в 4 раза и посмотрим, как это повлияет на f1-меру.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled
```

```python
features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```

### Downsampling

Хоть мы и решили поставленную задачу, давайте проверим сможет ли лучший наш алгоритм машинного обучения случайный лес добиться 0.6 по f1-мере на тестовой выборке, если мы уменьшим выборку для баланса классов.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
```

```python
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.4)
```

В целях экономии времени следующую ячейку можно не запускать.

```python
%%time
space = {
    'criterion': hp.choice('criterion', ['gini', 'entropy']),
    'max_features': hp.choice('max_features', [None, 'sqrt', 'log2']),
    'max_depth': hp.choice('max_depth', range(2, 50)),
    'min_samples_leaf': hp.choice('min_samples_leaf', range(2, 20)),
    'min_samples_split': hp.choice('min_samples_split',range(2, 20)),
    'n_estimators': hp.choice('n_estimators',range(10, 200))
}

def objective(params):
    clf = RandomForestClassifier(
        criterion=params['criterion'],
        max_depth=params['max_depth'],
        min_samples_split=params['min_samples_split'],
        min_samples_leaf=params['min_samples_leaf'],
        max_features=params['max_features'],
        n_estimators = params['n_estimators'],
        random_state=12345
    )

    clf.fit(features_downsampled, target_downsampled)

    y_pred = clf.predict(features_valid)

    score = f1_score(target_valid, y_pred)

    return {'loss': -score, 'params': params, 'status': STATUS_OK}

trials = Trials()

try:
    best = fmin(
        fn=objective,
        space=space,
        algo=tpe.suggest,
        max_evals=100,
        trials=trials,
        rstate=np.random.RandomState(12345))
except:
    best = fmin(
        fn=objective,
        space=space,
        algo=tpe.suggest,
        max_evals=100,
        trials=trials,
        rstate=np.random.default_rng(12345))

print("Best hyperparameters:")
print(best)

downsampled_rfc = RandomForestClassifier(
    max_depth=range(2, 50)[best['max_depth']],
    min_samples_split=range(2, 20)[best['min_samples_split']],
    min_samples_leaf=range(2, 20)[best['min_samples_leaf']],
    criterion=['gini', 'entropy'][best['criterion']],
    max_features=[None, 'sqrt', 'log2'][best['max_features']],
    n_estimators = range(10, 200)[best['n_estimators']],
    random_state=12345
)
downsampled_rfc.fit(features_downsampled, target_downsampled)
```

По сути градиентный бустинг это последовательное создание деревьев. Результаты предыдущего дерева влияют на следующие, как бы создание сбалансированного дерева, проверим, какой результат нам даст алгоритм градиентного бустинга. Следующую ячейку лучше не запускать.
